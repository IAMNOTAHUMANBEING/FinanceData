```python
import FinanceDataReader as fdr
import pandas as pd
from datetime import datetime
```

# 상장 + 폐지 기업 데이터

```python
# 기업 데이터 저장
stocks = fdr.StockListing('KRX')
krx_delisting = fdr.StockListing('KRX-DELISTING')
```

```python
# ETF, 콜풋 등 제거 필요
stocks.info()
krx_delisting.info()
```

```python
# 우선주 추출
prefer = stocks.loc[stocks.Name.str.contains(r'우$')]
prefer
```

```python
# 기업만 추출
Enterprise = stocks.dropna(subset=['Sector'])
Enterprise
```

```python
Enterprise = Enterprise[['Symbol','Name','Market']]
Enterprise
```

```python
prefer = prefer[['Symbol','Name','Market']]
prefer
```

```python
stock_list = pd.concat([Enterprise, prefer])
stock_list = stock_list.reset_index().drop('index', axis=1)
stock_list.to_csv('stock_list.csv', encoding='utf-8-sig') 
```

# 종목별 전체 주가 데이터(상장 + 폐지)

```python
for code in stock_list['Symbol']:
    data = fdr.DataReader(code)
    data = data.reset_index() 
    data['Date'] = data['Date'].apply(lambda x : datetime.strftime(x, '%Y-%m-%d'))
    data.to_csv('prices/' + code + '.csv', encoding='utf-8-sig')
```
